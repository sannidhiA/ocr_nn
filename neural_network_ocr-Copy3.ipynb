```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))
#transfer derivative
def dsigmoid(x):
    return x*(1-x)
```

```python
data=pd.read_csv("optdigits.tra",header=None)
data
```

```python
def feedforward(x,w0,w1,w2,w3):
    layer0=x
    layer1=sigmoid(np.dot(layer0,w0))
    layer2=sigmoid(np.dot(layer1,w1))
    layer3=sigmoid(np.dot(layer2,w2))
    layer4=sigmoid(np.dot(layer3,w3))
    return layer0,layer1,layer2,layer3,layer4#required for backprop
    
```

```python
def backwardpropagation(i,a0,a1,a2,a3,a4,w0,w1,w2,w3,y,lr):
    l4_error=a4-y
    l4_delta=l4_error*dsigmoid(a4)
    dh4=np.dot(a3.T,l4_delta)
    
    #l3_error=a3-y
    #l3_delta=l3_error*dsigmoid(a3)
    #dh3=np.dot(a2.T,l3_delta)
    
    l3_error=l4_delta.dot(w3.T)    
    l3_delta=l3_error*dsigmoid(a3)
    dh3=np.dot(a2.T,l3_delta)
    
    l2_error=l3_delta.dot(w2.T)    
    l2_delta=l2_error*dsigmoid(a2)
    dh2=np.dot(a1.T,l2_delta)
    
    l1_error=l2_delta.dot(w1.T)    
    l1_delta=l1_error*dsigmoid(a1)
    dh1=np.dot(a0.T,l1_delta)
    
    w3=w3-(dh4*lr)
    w2=w2-(dh3*lr)
    w1=w1-(dh2*lr)
    w0=w0-(dh1*lr)
    
    loss=np.mean(np.power(l4_error,2))
    loss_curve.append(loss)
    
    if i%1000 == 0:
        print(i,"error:",loss)
    return w0,w1,w2,w3
    
```

```python
w0=np.random.random((64,4))
w1=np.random.random((4,10))
w2=np.random.random((10,6))
w3=np.random.random((6,10))

```

```python
iteration=range(10000)
lr=0.01
loss_curve=[]
```

```python
for j in iteration:
    a0,a1,a2,a3,a4=feedforward(f,w0,w1,w2,w3)
    w0,w1,w2,w3=backwardpropagation(j,a0,a1,a2,a3,a4,w0,w1,w2,w3,target,lr)
```

```python
def accuracy(testx,testy):
    correct=0
    a0,a1,a2,a3,a4=feedforward(testx,w0,w1,w2,w3)
    for i in range(len(testx)):
        if np.argmax(a4[i])== np.argmax(testy[i]):
            correct+=1
    return (correct*100)/len(testx)
```
